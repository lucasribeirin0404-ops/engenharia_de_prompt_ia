```python
# Solicita ao usuário o primeiro número e o converte para float
num1 = float(input("Digite o primeiro número: "))

# Solicita ao usuário o segundo número e o converte para float
num2 = float(input("Digite o segundo número: "))

# Solicita ao usuário a operação desejada e converte para minúsculas para padronização
operacao = input("Escolha a operação (somar ou subtrair): ").lower()

# Verifica qual operação foi escolhida pelo usuário
if operacao == "somar":
    # Realiza a soma dos dois números
    resultado = num1 + num2
    # Exibe o resultado da soma
    print(f"A soma é: {resultado}")
elif operacao == "subtrair":
    # Realiza a subtração dos dois números
    resultado = num1 - num2
    # Exibe o resultado da subtração
    print(f"A subtração é: {resultado}")
else:
    # Caso a operação seja inválida, exibe uma mensagem de erro
    print("Operação inválida. Por favor, escolha 'somar' ou 'subtrair'.")
```
