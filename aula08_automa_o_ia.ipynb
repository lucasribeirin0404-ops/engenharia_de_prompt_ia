```python
# Define a variável current_temperature com um valor inicial para teste.
# Você pode alterar este valor para simular diferentes temperaturas.
current_temperature = 35

# Verifica se a temperatura atual é maior que 30 graus Celsius.
if current_temperature > 30:
    # Se a temperatura for maior que 30, exibe um alerta de superaquecimento.
    print("ALERTA: Temperatura atual é " + str(current_temperature) + "°C. Risco de superaquecimento!")
else:
    # Caso contrário (se a temperatura for 30 ou menos), exibe uma mensagem de sistema normal.
    print("Temperatura atual é " + str(current_temperature) + "°C. O sistema está normal.")
```
