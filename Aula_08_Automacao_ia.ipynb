```python
import os
import shutil

def organize_files_by_extension(directory_path):
    """
    Organiza arquivos no diretório especificado em subpastas baseadas em suas extensões,
    usando nomes de pasta em português.
    """
    if not os.path.isdir(directory_path):
        print(f"Erro: O diretório '{directory_path}' não existe.")
        return

    print(f"Iniciando organização no diretório: {directory_path}")

    # Dicionário para mapear extensões para nomes de pastas em português
    extension_to_portuguese_folder = {
        '.pdf': 'Documentos',
        '.doc': 'Documentos',
        '.docx': 'Documentos',
        '.txt': 'Textos',
        '.csv': 'Dados',
        '.xls': 'Planilhas',
        '.xlsx': 'Planilhas',
        '.ppt': 'Apresentacoes',
        '.pptx': 'Apresentacoes',
        '.jpg': 'Imagens',
        '.jpeg': 'Imagens',
        '.png': 'Imagens',
        '.gif': 'Imagens',
        '.bmp': 'Imagens',
        '.svg': 'Imagens',
        '.mp3': 'Audios',
        '.wav': 'Audios',
        '.ogg': 'Audios',
        '.mp4': 'Videos',
        '.avi': 'Videos',
        '.mov': 'Videos',
        '.mkv': 'Videos',
        '.zip': 'Compactados',
        '.rar': 'Compactados',
        '.7z': 'Compactados',
        '.exe': 'Executaveis',
        '.msi': 'Executaveis',
        '.py': 'Scripts',
        '.js': 'Scripts',
        '.html': 'Web',
        '.css': 'Web',
        '.json': 'Dados',
        '.xml': 'Dados'
        # Adicione mais mapeamentos conforme necessário
    }

    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)

        # Ignorar diretórios e arquivos ocultos ou de sistema
        if os.path.isfile(file_path) and not filename.startswith('.'):
            # Extrair a extensão do arquivo
            _, extension = os.path.splitext(filename)
            extension = extension.lower() # Converter para minúsculas para padronização

            if not extension: # Se não houver extensão, tratar como 'Outros'
                folder_name = "Outros"
            else:
                # Obter o nome da pasta em português, ou usar a extensão original sem o '.'
                folder_name = extension_to_portuguese_folder.get(extension, extension[1:])

            destination_folder = os.path.join(directory_path, folder_name)

            # Criar a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"Pasta '{folder_name}' criada.")

            # Mover o arquivo para a pasta de destino
            try:
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{folder_name}/'")
            except shutil.Error as e:
                print(f"Erro ao mover '{filename}': {e}")
        elif os.path.isdir(file_path) and filename not in extension_to_portuguese_folder.values() and filename != os.path.basename(directory_path) and filename != 'Outros':
             print(f"Ignorando subdiretório: {filename}")

    print("Organização concluída!")

# --- Como usar o programa ---
# 1. Altere o 'path_to_organize' para o caminho do diretório que você deseja organizar.
# 2. Execute esta célula.

# Exemplo de uso:
# Substitua 'C:/Users/SeuUsuario/Downloads' ou '/home/seu_usuario/Downloads' pelo caminho real.
# path_to_organize = "/caminho/do/seu/diretorio/baguncado"
# path_to_organize = "/content/files_to_organize" # Exemplo para Google Colab, se você subir arquivos

# Você pode descomentar a linha abaixo e fornecer o caminho.
# path_to_organize = input("Por favor, digite o caminho completo do diretório a ser organizado: ")

# Criando um diretório de teste e alguns arquivos para demonstração (opcional)
# Remova ou comente este bloco se você já tiver um diretório para organizar
test_dir = "./test_files_to_organize"
if not os.path.exists(test_dir):
    os.makedirs(test_dir)
    open(os.path.join(test_dir, "documento1.pdf"), "w").close()
    open(os.path.join(test_dir, "foto1.jpg"), "w").close()
    open(os.path.join(test_dir, "relatorio.docx"), "w").close()
    open(os.path.join(test_dir, "imagem.png"), "w").close()
    open(os.path.join(test_dir, "script.py"), "w").close()
    open(os.path.join(test_dir, "arquivo_sem_extensao"), "w").close()
    print(f"Diretório de teste '{test_dir}' criado com arquivos de exemplo.")

path_to_organize = test_dir # Use o diretório de teste para esta execução

organize_files_by_extension(path_to_organize)

# Para limpar o diretório de teste após a execução (opcional)
# descomente a linha abaixo se quiser remover os arquivos e pastas criados no teste
# shutil.rmtree(test_dir)
```
